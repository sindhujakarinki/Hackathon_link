```python
import cv2
import mediapipe as mp

mp_hands = mp.solutions.hands
hands = mp_hands.Hands()
mp_draw = mp.solutions.drawing_utils
cap = cv2.VideoCapture(0)

finger_tips = [8, 12, 16, 20]
thumb_tip = 4

while True:
    ret, img = cap.read()
    img = cv2.flip(img, 1)
    h, w, c = img.shape
    results = hands.process(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            lm_list = []
            for id, lm in enumerate(hand_landmarks.landmark):
                lm_list.append(lm)

            finger_fold_status = []
            for tip in finger_tips:
                x, y = int(lm_list[tip].x * w), int(lm_list[tip].y * h)
                if lm_list[tip].x < lm_list[tip - 2].x:
                    finger_fold_status.append(True)
                else:
                    finger_fold_status.append(False)

            print(finger_fold_status)

            x, y = int(lm_list[8].x * w), int(lm_list[8].y * h)
            print(x, y)

            gestures = {
                "STOP": all([
                    lm_list[4].y < lm_list[2].y,
                    lm_list[8].y < lm_list[6].y,
                    lm_list[12].y < lm_list[10].y,
                    lm_list[16].y < lm_list[14].y,
                    lm_list[20].y < lm_list[18].y,
                    lm_list[17].x < lm_list[0].x < lm_list[5].x
                ]),
                "FORWARD": all([
                    lm_list[3].x > lm_list[4].x,
                    lm_list[8].y < lm_list[6].y,
                    lm_list[12].y > lm_list[10].y,
                    lm_list[16].y > lm_list[14].y,
                    lm_list[20].y > lm_list[18].y
                ]),
                "BACKWARD": all([
                    lm_list[3].x > lm_list[4].x,
                    lm_list[3].y < lm_list[4].y,
                    lm_list[8].y > lm_list[6].y,
                    lm_list[12].y < lm_list[10].y,
                    lm_list[16].y < lm_list[14].y,
                    lm_list[20].y < lm_list[18].y
                ]),
                "LEFT": all([
                    lm_list[4].y < lm_list[2].y,
                    lm_list[8].x < lm_list[6].x,
                    lm_list[12].x > lm_list[10].x,
                    lm_list[16].x > lm_list[14].x,
                    lm_list[20].x > lm_list[18].x,
                    lm_list[5].x < lm_list[0].x,
                ]),
                "RIGHT": all([
                    lm_list[4].y < lm_list[2].y,
                    lm_list[8].x < lm_list[6].x,
                    lm_list[12].x > lm_list[10].x,
                    lm_list[16].x > lm_list[14].x,
                    lm_list[20].x > lm_list[18].x,
                    lm_list[5].x < lm_list[0].x,
                ]),
                "OK": all([
                    lm_list[thumb_tip].x < lm_list[thumb_tip - 1].x < lm_list[thumb_tip - 2].x,
                    lm_list[4].y < lm_list[3].y,
                    lm_list[8].y < lm_list[6].y,
                    lm_list[12].y < lm_list[10].y,
                    lm_list[16].y < lm_list[14].y,
                    lm_list[20].y < lm_list[18].y,
                ])
            }

            for gesture, condition in gestures.items():
                if condition:
                    cv2.putText(img, gesture, (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
                    print(gesture)

            if all(finger_fold_status):
                if (
                    lm_list[thumb_tip].y < lm_list[thumb_tip - 1].y < lm_list[thumb_tip - 2].y and
                    lm_list[0].x < lm_list[3].x and
                    lm_list[5].y < lm_list[4].y
                ):
                    gesture_text = "LIKE"
                    color = (0, 255, 0)
                elif (
                    lm_list[thumb_tip].y > lm_list[thumb_tip - 1].y > lm_list[thumb_tip - 2].y and
                    lm_list[0].x < lm_list[3].x and
                    lm_list[5].y > lm_list[4].y and
                    lm_list[8].y > lm_list[6].y
                ):
                    gesture_text = "DISLIKE"
                    color = (0, 0, 255)
                else:
                    gesture_text = ""
                    color = None

                if gesture_text:q
                    cv2.putText(img, gesture_text, (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 3)
                    print(gesture_text)

            mp_draw.draw_landmarks(
                img,
                hand_landmarks,
                mp_hands.HAND_CONNECTIONS,
                mp_draw.DrawingSpec((255, 255, 255), 6, 2),
                mp_draw.DrawingSpec((0, 0, 0), 4, 2)
            )

    cv2.imshow("Hand Sign Detection", img)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
